```python
import datasets
from bpe_tokenizer import BPECharacterTokenizer, BPEByteTokenizer
```

```python
dataset = datasets.load_dataset("ag_news")

# create a tokenizer training dataset
tokenizer_train = list(dataset["train"]["text"][:])
```

```python
char_tokenizer = BPECharacterTokenizer()
char_tokenizer.train(tokenizer_train, 1000, True)
```

```python
print(len(char_tokenizer.vocabulary))
print(char_tokenizer.vocabulary)
```

```python
char_tokenizer.encode("Hello")
```

```python
char_tokenizer.decode([39, 368, 74, 0])
```

```python
byt_tokenizer = BPEByteTokenizer()
byt_tokenizer.train(tokenizer_train, 1000)
```

```python
print(len(byt_tokenizer.vocabulary))
print(byt_tokenizer.vocabulary)
```

```python
byt_tokenizer.encode("Hello")
```

```python
byt_tokenizer.decode([76, 542, 115, 199, 156])
```

```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import ByteLevel, Split
from tokenizers.decoders import ByteLevel as ByteLevelDecoder, BPEDecoder


tokenizer = Tokenizer(BPE(unk_token="<unk>"))
tokenizer.decoder = BPEDecoder()
trainer = BpeTrainer(
    vocab_size=1000,
    special_tokens=["<pad>", "<unk>", "<bos>", "<eos>"],
    show_progress=True,
)

tokenizer.train_from_iterator(tokenizer_train, trainer)
```

```python
tokenizer.encode("Hello").tokens
```

```python
tokenizer.decode(tokenizer.encode("Hello").ids)
```

```python
byte_tokenizer = Tokenizer(BPE(unk_token="<unk>"))
byte_tokenizer.pre_tokenizer = ByteLevel(add_prefix_space=False)
byte_tokenizer.decoder = ByteLevelDecoder()


trainer = BpeTrainer(
    vocab_size=1000,
    special_tokens=["<pad>", "<unk>", "<bos>", "<eos>"],
    initial_alphabet=ByteLevel.alphabet(),
    show_progress=True,
)

byte_tokenizer.train_from_iterator(tokenizer_train, trainer)
```

```python
byte_tokenizer.decode(byte_tokenizer.encode("Hello").ids)
```
